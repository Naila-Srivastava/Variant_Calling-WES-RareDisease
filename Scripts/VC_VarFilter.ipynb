```python
## Import the necessary packages

import pandas as pd
import matplotlib.pyplot as plt

## Load the variant data
df = pd.read_csv('SRR18395025_final_variants.txt', sep='\t')
print(df)
```

```python
## Count unique variants per gene
gene_counts = df.groupby('SYMBOL')['Variant'].nunique().sort_values(ascending=False)

## Plot top 10 genes by number of unique variants
plt.figure(figsize=(10,6))
gene_counts.head(10).plot(kind='bar')
plt.xlabel('Gene')
plt.ylabel('Number of Unique Variants')
plt.title('Top 10 Genes by Unique Variant Count')
plt.tight_layout()

## Show and save
plt.savefig("top_genes_variants.png")
plt.show()
```

```python
## 1. Variants per gene
gene_counts = df.groupby('SYMBOL')['Variant'].nunique().sort_values(ascending=False)
plt.figure(figsize=(10,4))
gene_counts.plot(kind='bar')
plt.ylabel('Number of unique variants')
plt.title('Variants per Gene')
plt.tight_layout()
plt.savefig("variants_per_gene.png")
plt.show()


## 2. SIFT categories
sift = df['SIFT'].dropna().str.extract(r'([a-z_]+)', expand=False).value_counts()
plt.figure()
sift.plot(kind='bar')
plt.ylabel('Number of Variants')
plt.title('SIFT Categories')
plt.tight_layout()
plt.savefig("sift_categories.png")
plt.show()

# 3. PolyPhen categories
polyphen = df['PolyPhen'].dropna().str.extract(r'([a-z_]+)', expand=False).value_counts()
plt.figure()
polyphen.plot(kind='bar')
plt.ylabel('Number of Variants')
plt.title('PolyPhen Categories')
plt.tight_layout()
plt.savefig("polyphen_categories.png")
plt.show()
```
